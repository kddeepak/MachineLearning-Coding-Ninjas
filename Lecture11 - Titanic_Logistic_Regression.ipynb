```python
from sklearn.linear_model import LogisticRegression
import numpy as np
import pandas as pd
```

```python
def run():
    df=pd.read_csv("0000000000002429_training_titanic_x_y_train.csv") 
    df2=pd.read_csv("0000000000002429_test_titanic_x_test.csv") 
    #test_data = np.genfromtxt("0000000000002429_test_titanic_x_test.csv",delimiter=',')
    #test_data = df2.values
    del df['Name']
    del df['Ticket']
    del df['Fare']
    del df['Cabin']
    
    del df2['Name']
    del df2['Ticket']
    del df2['Fare']
    del df2['Cabin']
    
    def getNumber(str):
        if str=="male":
            return 0
        else:
            return 1
    
    df["Gender"]=df["Sex"].apply(getNumber)
    df2["Gender"]=df2["Sex"].apply(getNumber)
    
    def fun(s):
        if s=='C':
            return 0
        elif s=='Q':
            return 1
        else:
            return 2

    df2['Embarked'] = df2.Embarked.apply(fun)
    df['Embarked'] = df.Embarked.apply(fun)
    
    del df['Sex']
    del df2['Sex']
    
    #filling the nan values for the age attribute
    meanS= df[df.Survived==1].Age.mean()
    meanS2= df2.Age.mean()
    
    df["age"]=np.where(pd.isnull(df.Age) & df["Survived"]==1  ,meanS, df["Age"])
    df2["age"]=np.where(pd.isnull(df2.Age),meanS2, df2["Age"])
    
    meanNS=df[df.Survived==0].Age.mean()
    #meanNS2=df2[df2.Survived==0].Age.mean()

    
    df.age.fillna(meanNS,inplace=True)
    #df2.age.fillna(meanNS2,inplace=True)
    
    #print()
    del df['Age']
    del df2['Age']
    
    df['gender'] = df['Gender']
    df['survived'] = df['Survived']
    df2['gender'] = df2['Gender']
    #df2['survived'] = df2['Survived']
    del df['Gender']
    del df['Survived']
    
    del df2['Gender']
    #del df2['Survived']
    print(df.head())
    print(df2.head())
    
    data = df.values
    data_test = df2.values
   # print(df.head())
    train_x = data[:,0:6]
    train_y = data[:,6]
    
   # print(train_x)
    alg = LogisticRegression()
    alg.fit(train_x,train_y)
    
    prediction_titanic = alg.predict(data_test)
    print(prediction_titanic)
    np.savetxt("titanic_prediction.csv",prediction_titanic,delimiter=',')
    #y_train = train_data[:,10]
    
```

```python
run()
```
