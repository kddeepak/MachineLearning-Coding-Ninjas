```python
#need to add the bias term in the code
import pandas as pd
import numpy as np
from sklearn import datasets
from math import exp

```

```python
def sigmoid(z):
    
    temp = exp(z)
    return temp/(1+temp)

def predict(data,m):
    temp = np.dot(data,m)
    print(temp)
    for i in range(len(temp)):
        temp[i]= sigmoid(temp[i])
        if temp[i]<0.5:
            temp[i]=0
        else:
            temp[i]=1
    return temp
    
    
```

```python
def step_gradient(data,y_train,alpha,m):
    m_slope = np.zeros(len(data[0]))
    
    for i in range(len(data[0])):
        for j in range(len(data)):
            sum=0.0
            for k in range(len(data[0])):
                sum+=(m[k]*data[j,k])
            m_slope[i]+=(-1/len(data))*(y_train[j]-sigmoid(sum))*data[j,i]  
            
    for i in range(len(m_slope)):
        m[i]=m[i]-alpha*m_slope[i]
    return m    
    
    
```

```python
def lg(data,y_train,iterations,alpha):
    m = np.zeros(len(data[0]))
    for i in range(iterations):
        m = step_gradient(data,y_train,alpha,m)
    return m


    
```
