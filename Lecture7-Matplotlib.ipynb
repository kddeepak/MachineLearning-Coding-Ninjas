# MatplotLib

```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3]
y=[2,4,6]
```

```python
plt.scatter(x,y)
plt.show()
```

```python
x = [1,2,3]
y=[2,4,6]

x2=[3,4,5]
y2=[3,4,8]
plt.scatter(x,y)
plt.scatter(x2,y2)
plt.plot(x,y,"o-.")
plt.plot(x2,y2,"r--")
plt.show()
```

```python
#plot function x^3
import numpy as np
#x= np.array([1,2,3,4])
x = np.arange(0,5,0.1)
y = x**2

plt.plot(x,y,"b.")
```

```python
#color ,marker ,line width,label
x = np.arange(0,5,0.1)

y = x**3
y2 = x**2
plt.plot(x,y,color='black',linewidth=5,label="x^3")
plt.plot(x,y2,color='red',linewidth=5,label="x^2")
plt.axis([0,10,0,100])
plt.ylabel("y")
plt.xlabel("X")
plt.grid()

plt.text(1,20,"text")
plt.legend()#to show legends
plt.title("Matplotlib demo")
plt.show()
```

```python
#drawing a pie chart
sizes =[3,4,6,2]
color =["red","blue","purple","yellow"]
labels = ["A","B","C","D"]
explode =[0.1,0.2,0,0]
plt.pie(sizes,colors=color,labels=labels,autopct="%.1f%%",explode =explode,counterclock =False,startangle=90
       )#double % required to print a %.
plt.title("Split among classes")
plt.axis("equal")
plt.show()
```
