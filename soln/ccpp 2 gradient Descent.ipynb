```python
import numpy as np
import pandas as pd
import csv
from sklearn import preprocessing

def step_gradient(data,y_train,learning_rate,m):
    x=data #13 columns
    y=y_train
    #Converting x to a dataframe df for adding a new column.
    df1=pd.DataFrame(x)
    df1['s7']=1 #Added a 14th column with all 1s in place of 'c'.
    #print(df1.describe())
    #And then reconverting back into a list.
    x1=df1.values
    m_slope=np.zeros(5)
    M=len(data)
    for i in range(M):
        for j in range(5):
            sum=0
            for k in range(5):
                sum+=m[k]*x1[i,k]
            m_slope[j]+=(-2/M)*(y[i]-sum)*x1[i,j]
            
    for i in range(5):
        m[i]=m[i]-learning_rate*m_slope[i]
    return m
        
    
    
def calculate_cost(data,y_train,m):
    x=data
    y=y_train
    df1=pd.DataFrame(x)
    df1['s7']=1
    x1=df1.values
    total_cost=0
    M=len(data)
    for i in range(M):
        sum=0
        for j in range(5):
            sum+=m[j]*x1[i,j]
        total_cost+=(1/M)*((y[i]-sum)**2)
    return total_cost

def gd(data,y_train,learning_rate,num_iterations):
    m=np.zeros(5)
    for i in range(num_iterations):
        m=step_gradient(data,y_train,learning_rate,m)
        cost=calculate_cost(data,y_train,m)
        print(cost)
    return m

def predict(data1,m):
    df=pd.DataFrame(data1)
    
    df['c']=1
    x1=df.values
    y_pred=np.dot(x1,m)
   
    return y_pred
    
    
def run():
    data=np.genfromtxt("0000000000002419_training_ccpp_x_y_train.csv", delimiter=",")
    Y_train = data[:,4]
    datacopy = data[:,0:4]
    learning_rate=0.1
    num_iterations=200
    scaler=preprocessing.StandardScaler().fit(datacopy)
    X_train_scaled=scaler.transform(datacopy)
    
    m=gd(X_train_scaled,Y_train,learning_rate,num_iterations)
    data1=np.genfromtxt("0000000000002419_test_ccpp_x_test.csv", delimiter=",")
    X_test_scaled=scaler.transform(data1)
    y_pred=predict(X_test_scaled,m)
    y_pred=np.around(y_pred,decimals=5)
    #print(y_pred)
    np.savetxt("predictions_of_ccpp.csv",y_pred,delimiter=",")
    
run()
```
